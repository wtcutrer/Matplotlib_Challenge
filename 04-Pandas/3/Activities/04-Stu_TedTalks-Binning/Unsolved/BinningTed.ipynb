```python
# Import Dependencies
import pandas as pd
```

```python
# Create a path to the csv and read it into a Pandas DataFrame
csv_path = "Resources/ted_talks.csv"
df = pd.read_csv(csv_path)

df.head()
```

```python
# Create bins in which to place values based upon TED Talk views
bins = [0, 199999, 399999, 599999, 799999, 999999, 1999999, 2999999, 3999999, 4999999, 50000000]

# Create labels for these bins

group_names = ["Sucks" , "Kinda sucks" , "sucks less" , "not bad" , "kinda okay",
               "okay" , "kinda good" , "pretty good" , "great" , "amazing"]
```

```python
print(len(bins), len(group_names))
```

```python
# Slice the data and place it into bins
df["ted_talks"] = pd.cut(df["views"], bins, labels=group_names, include_lowest=True)
df
```
